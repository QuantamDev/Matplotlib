```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0,1,2,3,4]
y = [0,1,2,3,4]

plt.plot(x,y, label='2x',color='blue',linewidth=2,marker='.',markersize=10,linestyle='dotted')
plt.title('My first graph',{'fontsize':20})
plt.xlabel('age')
plt.ylabel('BMI')

#second line
# x2=np.arrange(0,4.5,0.5)
# plt.plot(x2,x2**2)

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```
