```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0,1,2,3,4]
y = [0,1,2,3,4]

#Resize the graph
plt.figure(figsize=(5,3),dpi=100)


plt.plot(x,y, label='2x',color='blue',linewidth=2,marker='.',markersize=10,linestyle='dotted')
plt.title('My first line graph',{'fontsize':20})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


#second line
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.xticks([0,1,2,3,4,])
# plt.yticks([0,2,4,6,8,10])
plt.legend()

plt.savefig('mygraph.png',dpi=200)
plt.show()
```

#Bar Chart


```python
labels = ['A','B','C']
values =[1,4,2]
plt.title('My First Bar Graph')
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('+')
bars[2].set_hatch('-')
plt.figure(figsize=(3,5))
plt.show()
```
