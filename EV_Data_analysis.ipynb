```python
sns.countplot(x='Model',data=ev)
```

```python
sns.countplot(x='Model Year',data=ev)
sns.set(rc={figure.figsize:(10,5)})
```

```python
ev_data = ev.groupby(['Electric Vehicle Type'], as_index=False)['Base MSRP'].sum().sort_values(by='Base MSRP', ascending=False)

sns.barplot(x='Electric Vehicle Type', y='Base MSRP', data=ev_data)
```

```python
ev_range = ev.groupby(['Model'], as_index=False)['Electric Vehicle Type'].sum().sort_values(by='Electric Vehicle Type', ascending=False)
sns.barplot(x='Model', y='Electric Vehicle Type', data=ev_range)
```

```python
sns.countplot(data=ev,x='Model',hue='Electric Vehicle Type')
sns.set(rc={'figure.figsize':(15,5)})
```

### Reln between the Model and Model Year###

```python
model_data=ev.groupby(['Model Year'],as_index=False)['Model'].sum().sort_values(by='Model',ascending=False)
sns.set(rc={'figure.figsize':(55,40)})
sns.barplot(data=ev,x="Model Year",y="Model")
```

```python
model_data=ev.groupby(['Electric Vehicle Type'],as_index=False)['Model'].sum().sort_values(by='Model',ascending=False)
sns.set(rc={'figure.figsize':(55,40)})
sns.barplot(data=ev,x="Electric Vehicle Type",y="Model")
```
